```python
import json
import pandas as pd
import numpy as np
import networkx as nx
import itertools
```

```python
domain = open('data/adult.domain', 'r')
counts = json.load(open('data/adult-domain.json'))
```

```python
temp = []
for line in domain.readlines():
    this_line = line.replace("{", "")
    this_line = this_line.replace("}", "")
    this_line = this_line.strip("\n").split()
    temp.append(this_line)
```

```python
domain_info = dict()
for i,key in enumerate(counts.keys()):
    domain_info[key] = temp[i]
```

```python
possible_values = dict()
specs = dict()
mapping = dict()
```

```python
for key in counts.keys():
    specs[key] = dict()
    mapping[key] = dict()
    possible_values[key] = []
    dtype = domain_info[key][0]
    if dtype == "C":
        specs[key]["type"] = "numeric"
        lower = int(float(domain_info[key][1]))
        upper = int(float(domain_info[key][2]))
        specs[key]["lower_bound"] = lower
        specs[key]["upper_bound"] = upper
        nbins = int(counts[key])
        specs[key]["num_bins"] = nbins
        possible_values[key] = np.linspace(lower, upper, nbins)
    else:
        specs[key]["type"] = "character"
        specs[key]["num_bins"] = int(counts[key])
        possible_values[key] = domain_info[key][1:]
    
```

```python
for key in possible_values.keys():
    #print(possible_values[key])
    for i,value in enumerate(possible_values[key]):
        mapping[key][i] = possible_values[key][i]
```

```python
json.dump(mapping, open("data/mapping.json",'w'))
json.dump(specs, open("data/specs.json",'w'))
json.dump(possible_values, open("data/domain.json", 'w'))
```
